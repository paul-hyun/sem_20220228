# Matrix Multiplication

```
# B 행렬 matrix multiplication
B = np.random.randint(-9, 10, (4, 3)) / 10

print(A)
print()
print(B)
print()
print(np.matmul(A, B))
```

```
# tensorflow matmul
tf.matmul(A, B)
```

```
# 직접계산
result = np.zeros((A.shape[0], B.shape[-1]))
for row in range(A.shape[0]):
    for col in range(B.shape[-1]):
        result[row][col] = np.dot(A[row], B[:, col])

print(result)
```
